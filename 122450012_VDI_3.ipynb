Dataset yang akan digunakan dalam tugas ini yaitu dataset berisi tentang distribusi jabatan manager menurut jenis kelamin pada tahun 2016-2019. Data ini berasal dari BPS Indonesia.

1. Tujuan: untuk menunjukkan bahwa seiring berjalannya waktu, kesetaraan gender dalam bentuk kesamaan kesempatan setiap orang untuk menduduki suatu jabatan manajerial tanpa memandang gender semakin meningkat.
2. User: visualisasi ini ditujukan untuk semua orang secara global.
3. Fungsi: explanatory (menjelaskan dengan visual dari data yang sudah ada).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import dataset
data = pd.read_csv('/content/Distribusi Jabatan Manager Menurut Jenis Kelamin, 2018-2019.csv')
data
```

```python
# Mengubah dataset menjadi dataframe agar bisa dilakukan transformasi

data1 = pd.DataFrame(data)
data1
```

```python
# Mengubah dataframe ke format long untuk memudahkan visualisasi
df_long = pd.melt(data1, id_vars='Jenis Kelamin', var_name='Tahun', value_name='Persentase')

# Membuat barchart
plt.figure(figsize=(10, 6))
sns.barplot(x='Tahun', y='Persentase', hue='Jenis Kelamin', data=df_long, palette='Set2')

# Menambahkan judul dan label
plt.title('Distribusi Jabatan Manajer Menurut Jenis Kelamin (2016-2019)', fontsize=16)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Persentase (%)', fontsize=12)
plt.ylim(0, 100)

# Menampilkan plot
plt.show()
```
